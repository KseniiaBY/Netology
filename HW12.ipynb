```python
import pandas as pd
```

```python
data=pd.read_csv('movies.csv')
data.head()
```

```python
rat=pd.read_csv('ratings.csv')
rat.head()
```

```python
rat1=rat.groupby('movieId')['rating'].mean().reset_index()
rat1.head()
```

```python
data1=pd.merge(data,rat1, on='movieId')
data1
```

```python
def ratings(rating):
    if rating<=2:
        return 'низкий рейтинг'
    elif 2<=rating<=4:
        return 'средний рейтинг'
    elif rating==4.5 or rating==5:
         return 'высокий рейтинг'
```

```python
data1['class']=data1['rating'].apply(ratings)
data1
```

```python
df=pd.read_csv('keywords.csv')
df.head()
```

```python
def poisk(gorod):
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    c='undefined'
    for key,cities in geo_data.items():
        for city in cities:
            if city in gorod:
                c=key
    return c
```

```python
df['region']=df['keyword'].apply(poisk)
df['region'].value_counts()
```
